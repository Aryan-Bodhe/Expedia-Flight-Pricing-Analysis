```python
import pandas as pd
import numpy as np

# Ensure df is defined from your original data X

# ==============================================================================
# 1. DATETIME & TEMPORAL FEATURES
# ==============================================================================
df['searchDate'] = pd.to_datetime(df['searchDate'])
df['flightDate'] = pd.to_datetime(df['flightDate'])

df['flight_month'] = df['flightDate'].dt.month
df['flight_day_of_week'] = df['flightDate'].dt.dayofweek
df['is_weekend'] = df['flight_day_of_week'].apply(lambda x: 1 if x >= 5 else 0)
df['is_holiday_season'] = df['flight_month'].isin([7, 11, 12]).astype(int)

df['flight_month_sin'] = np.sin(2 * np.pi * df['flight_month'] / 12.0)
df['flight_month_cos'] = np.cos(2 * np.pi * df['flight_month'] / 12.0)

# ==============================================================================
# 2. SEGMENT COLUMN EXTRACTION (RE-FIXED FOR YOUR FORMAT)
# ==============================================================================
# Robust extraction function to handle '2022-05-15T01:35:00.000000000' and '||'
def get_first_hour(val):
    try:
        # Convert to string, take the first segment before '||', 
        # then split by 'T' and take the first 2 characters of the time part
        return float(str(val).split('||')[0].split('T')[1][:2])
    except:
        return np.nan

df['journey_start_hour'] = df['segmentsDepartureTime'].apply(get_first_hour)

# Extract Primary Airline
df['primary_airline'] = df['segmentsAirlineCode'].astype(str).str.split('||').str[0]
df['primary_airline'] = df['primary_airline'].replace('nan', np.nan)

# Check for mixed itineraries
df['is_mixed_itinerary'] = df['segmentsAirlineCode'].astype(str).apply(
    lambda x: 1 if len(set(x.split('||'))) > 1 else 0
)

# One-Hot Encode
df = pd.get_dummies(df, columns=['primary_airline'], prefix='airline', drop_first=False)

# ==============================================================================
# 3. TIME-WINDOWED TARGET ENCODING
# ==============================================================================
df['days_to_flight_bin'] = pd.cut(df['numDaysToFlight'], 
                                  bins=[0, 7, 14, 30, 60, 365],
                                  labels=['0-7', '8-14', '15-30', '31-60', '61+'],
                                  include_lowest=True)

cols_to_overwrite = [
    'start_airport_month_avg_fare', 'start_airport_month_count',
    'start_airport_daysbin_avg_fare', 'start_airport_daysbin_count',
    'dest_airport_month_avg_fare', 'dest_airport_month_count',
    'dest_airport_daysbin_avg_fare', 'dest_airport_daysbin_count'
]
df = df.drop(columns=[c for c in cols_to_overwrite if c in df.columns], errors='ignore')

if 'startingAirport' in df.columns:
    airport_month_targets = df.groupby(['startingAirport', 'flight_month'], observed=True)['totalFare'].agg(['mean', 'count']).reset_index()
    airport_month_targets.columns = ['startingAirport', 'flight_month', 'start_airport_month_avg_fare', 'start_airport_month_count']
    df = df.merge(airport_month_targets, on=['startingAirport', 'flight_month'], how='left')
    
    airport_days_targets = df.groupby(['startingAirport', 'days_to_flight_bin'], observed=True)['totalFare'].agg(['mean', 'count']).reset_index()
    airport_days_targets.columns = ['startingAirport', 'days_to_flight_bin', 'start_airport_daysbin_avg_fare', 'start_airport_daysbin_count']
    df = df.merge(airport_days_targets, on=['startingAirport', 'days_to_flight_bin'], how='left')

if 'destinationAirport' in df.columns:
    dest_month_targets = df.groupby(['destinationAirport', 'flight_month'], observed=True)['totalFare'].agg(['mean', 'count']).reset_index()
    dest_month_targets.columns = ['destinationAirport', 'flight_month', 'dest_airport_month_avg_fare', 'dest_airport_month_count']
    df = df.merge(dest_month_targets, on=['destinationAirport', 'flight_month'], how='left')
    
    dest_days_targets = df.groupby(['destinationAirport', 'days_to_flight_bin'], observed=True)['totalFare'].agg(['mean', 'count']).reset_index()
    dest_days_targets.columns = ['destinationAirport', 'days_to_flight_bin', 'dest_airport_daysbin_avg_fare', 'dest_airport_daysbin_count']
    df = df.merge(dest_days_targets, on=['destinationAirport', 'days_to_flight_bin'], how='left')

# ==============================================================================
# 4. FINAL CLEANUP
# ==============================================================================
y = df['totalFare'].copy()

drop_cols = [
    'legId', 'searchDate', 'flightDate', 'segmentsDepartureTime', 'segmentsArrivalTime',
    'segmentsAirlineName', 'segmentsAirlineCode', 'segmentsAircraft',
    'segmentsArrivalAirportCode', 'segmentsDepartureAirportCode',
    'startingAirport', 'destinationAirport', 'segmentsCabinCode',
    'baseFare', 'totalFare', 'flight_month', 'flight_day_of_week', 'days_to_flight_bin'
]

X_clean = df.drop(columns=[c for c in drop_cols if c in df.columns], errors='ignore')

# Handle types and NaNs
bool_cols = X_clean.select_dtypes(include='bool').columns
X_clean[bool_cols] = X_clean[bool_cols].astype(int)

X_numeric = X_clean.select_dtypes(include=['number'])
X_numeric = X_numeric.replace([np.inf, -np.inf], np.nan)
X_final = X_numeric.fillna(X_numeric.median(numeric_only=True))

print("PIPELINE COMPLETE")
print(f"Features: {X_final.shape[1]} | Rows: {X_final.shape[0]}")
```

```python
# Ensure X_final contains the necessary columns for interaction terms
# If a column is missing, we'll assign 0 or a reasonable default for calculation.

# 1. Demand Pressure: Interaction between booking window and remaining seats
# Lower numDaysToFlight means closer to departure, potentially higher price if seatsRemaining is low.
X_final['demand_pressure'] = (1 / (X_final['numDaysToFlight'] + 1)) * (1 / (X_final['seatsRemaining'] + 1))

# 2. Layover Discomfort Cost: Interaction of layover duration and total travel duration
# The perceived 'cost' of a layover might increase if the total travel duration is also long
X_final['layover_discomfort_cost'] = X_final['layoverDurationMinutes'] * X_final['travelDuration']

# 3. Seasonal Demand Impact: Interaction of monthly average fare and holiday season
# How much the holiday season amplifies or dampens the usual monthly average fare for an airport
X_final['seasonal_demand_impact'] = X_final['start_airport_month_avg_fare'] * X_final['is_holiday_season']

print("Synergistic features added to X_final.")
```

```python
# Calculate total minutes from midnight for departure
departure_minutes_from_midnight = X_final['journey_start_hour'] * 60

# Calculate raw arrival minutes from midnight
# travelDuration is in minutes
arrival_minutes_raw = departure_minutes_from_midnight + X_final['travelDuration']

# Normalize arrival_minutes to be within a single day (0 to 23*60 + 59)
# Use modulo 24 * 60 to handle flights that span multiple days.
X_final['arrival_minutes_of_day'] = arrival_minutes_raw % (24 * 60)

# Convert arrival_minutes_of_day back to hour
X_final['arrival_hour'] = X_final['arrival_minutes_of_day'] // 60

# Define what constitutes "late night arrival"
# User can adjust this range (e.g., 0 to 5 for 00:00 to 04:59)
late_night_start = 0
late_night_end = 5 # Exclusive, so includes 0, 1, 2, 3, 4

X_final['is_late_night_arrival'] = ((X_final['arrival_hour'] >= late_night_start) &
                                    (X_final['arrival_hour'] <= late_night_end)).astype(int)

# Create the interaction feature: travelDuration * is_late_night_arrival
X_final['late_night_arrival_impact'] = X_final['is_late_night_arrival'] * X_final['travelDuration']

print("New synergy features related to arrival time added to X_final:")
print("  - arrival_hour")
print("  - is_late_night_arrival")
print("  - late_night_arrival_impact")
```

```python

# 1. Travel Efficiency: travelDuration / totalTravelDistance
X_final['travel_efficiency'] = X_final['travelDuration'] / (X_final['totalTravelDistance'] + 1)

# 2. Layover Proportion: layoverDurationMinutes / travelDuration
X_final['layover_proportion'] = X_final['layoverDurationMinutes'] / (X_final['travelDuration'] + 1)

# 3. Non-stop Premium under Demand: isNonStop * demand_pressure
X_final['nonstop_premium_demand'] = X_final['isNonStop'] * X_final['demand_pressure']
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 1. Define your final columns list
final_cols = [
    'numDaysToFlight', 'numLayovers', 'layoverDurationMinutes', 'seatsRemaining', 
    'travelDuration', 'totalTravelDistance', 'elapsedDays', 'isBasicEconomy', 
    'isRefundable', 'isNonStop', 'is_weekend', 'is_holiday_season', 
    'flight_month_sin', 'flight_month_cos', 'journey_start_hour', 
    'is_mixed_itinerary', 'airline_', 'start_airport_month_avg_fare', 
    'start_airport_month_count', 'start_airport_daysbin_avg_fare', 
    'start_airport_daysbin_count', 'dest_airport_month_avg_fare', 
    'dest_airport_month_count', 'dest_airport_daysbin_avg_fare', 
    'dest_airport_daysbin_count', 'demand_pressure', 'layover_discomfort_cost', 
    'seasonal_demand_impact', 'arrival_minutes_of_day', 'arrival_hour', 
    'is_late_night_arrival', 'late_night_arrival_impact', 'travel_efficiency', 
    'layover_proportion', 'nonstop_premium_demand'
]

# Ensure you only take columns that actually exist in your dataframe
X = X_final[final_cols]

# 2. Standardize the data (CRITICAL for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Initial PCA to see Explained Variance
# We don't limit components yet to see the "Scree Plot"
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# 4. Determine how many components to keep
explained_variance = np.cumsum(pca.explained_variance_ratio_)

plt.figure(figsize=(10, 5))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--')
plt.title('Explained Variance by Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.axhline(y=0.95, color='r', linestyle='-')
plt.text(0.5, 0.85, '95% cut-off threshold', color = 'red', fontsize=12)
plt.grid()
plt.show()

# 5. Apply PCA with the optimal number of components (e.g., enough to cover 95% variance)
pca_final = PCA(n_components=0.95) 
X_reduced = pca_final.fit_transform(X_scaled)

print(f"Original feature count: {X.shape[1]}")
print(f"Reduced feature count: {X_reduced.shape[1]}")
```

```python
# Randomized PCA on the same feature set
X = X_final[final_cols]

# Standardize
X_scaled = scaler.fit_transform(X)  # reusing existing scaler object

# Fit full randomized PCA (to inspect cumulative variance)
pca_rand_full = PCA(svd_solver="randomized", random_state=42)
pca_rand_full.fit(X_scaled)

cum_var = np.cumsum(pca_rand_full.explained_variance_ratio_)
n_comp_95 = np.argmax(cum_var >= 0.95) + 1

print(f"Components needed for 95% variance (randomized PCA): {n_comp_95}/{X.shape[1]}")
print(f"Cumulative explained variance at cutoff: {cum_var[n_comp_95-1]:.4f}")

# Fit final randomized PCA with the integer component count
pca_rand_95 = PCA(n_components=n_comp_95, svd_solver="randomized", random_state=42)
X_reduced_rand = pca_rand_95.fit_transform(X_scaled)

# -------------------------------------------------------------------
# Feature contribution scoring (heuristic):
# weighted squared loading per original column across retained PCs
# -------------------------------------------------------------------
loadings = pca_rand_95.components_.T  # shape: (n_features, n_components_retained)
weights = pca_rand_95.explained_variance_ratio_  # per retained PC
feature_scores = (loadings**2 @ weights)  # contribution score per original feature

contrib_df = pd.DataFrame({
    "feature": X.columns,
    "score": feature_scores
}).sort_values("score", ascending=False).reset_index(drop=True)

contrib_df["cum_score"] = contrib_df["score"].cumsum() / contrib_df["score"].sum()

# Columns retained/removed by 95% contribution mass
retained_cols = contrib_df.loc[contrib_df["cum_score"] <= 0.95, "feature"].tolist()
if len(retained_cols) < len(contrib_df):  # include first feature that crosses 95%
    retained_cols.append(contrib_df.loc[len(retained_cols), "feature"])

removed_cols = [c for c in X.columns if c not in retained_cols]

print(f"\nRetained original columns (95% contribution): {len(retained_cols)}")
print(retained_cols)

print(f"\nRemoved original columns ({len(removed_cols)}):")
print(removed_cols)

# Optional: view contribution table
contrib_df.head(20)
```

```python
# Step 1: Baseline Multiple Linear Regression (MLR) using 32 predictors
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from statsmodels.stats.diagnostic import het_breuschpagan

# Remove the 3 low-contribution columns from the 35-feature set
cols_to_remove = ['is_mixed_itinerary', 'airline_', 'demand_pressure']
X_32 = X_final.drop(columns=cols_to_remove, errors='ignore').copy()

print(f"Initial features in X_final: {X_final.shape[1]}")
print(f"Features after dropping 3 columns: {X_32.shape[1]}")
print("Dropped columns:", cols_to_remove)

# Prepare clean regression frame
mlr_df = X_32.copy()
mlr_df['totalFare'] = y
mlr_df = mlr_df.replace([np.inf, -np.inf], np.nan).dropna()

X_mlr = mlr_df.drop(columns=['totalFare'])
y_mlr = mlr_df['totalFare']

# Fit OLS baseline
X_mlr_const = sm.add_constant(X_mlr, has_constant='add')
ols_model = sm.OLS(y_mlr, X_mlr_const).fit()

# Save fitted values and residuals for diagnostics in next cells
y_hat = ols_model.fittedvalues
resid = ols_model.resid
std_resid = ols_model.get_influence().resid_studentized_internal

print(f"\nObservations used in OLS: {len(y_mlr):,}")
print(f"Predictors used in OLS: {X_mlr.shape[1]}")
print(f"Training R^2: {r2_score(y_mlr, y_hat):.4f}")
print(f"Training RMSE: {np.sqrt(mean_squared_error(y_mlr, y_hat)):.4f}")
print(f"Training MAE: {mean_absolute_error(y_mlr, y_hat):.4f}")

print("\nModel summary (top section):")
display(ols_model.summary().tables[0])

coef_df = ols_model.summary2().tables[1].sort_values('P>|t|')
print("Top 10 most statistically significant coefficients:")
display(coef_df.head(10))

# ---------------------------------------------------------
# RESIDUAL DIAGNOSTICS (Memory Safe)
# ---------------------------------------------------------
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 1) Residuals vs Fitted (non-linearity and funnel check)
axes[0].scatter(y_hat, resid, alpha=0.2, s=10)
axes[0].axhline(0, color='red', linestyle='--', linewidth=1)

# Quadratic trendline to detect U-shape pattern
x_grid = np.linspace(y_hat.min(), y_hat.max(), 300)
quad_coef = np.polyfit(y_hat, resid, deg=2)
quad_trend = np.polyval(quad_coef, x_grid)
axes[0].plot(x_grid, quad_trend, color='black', linewidth=2, label='Quadratic trend')
axes[0].set_title('Residuals vs Fitted')
axes[0].set_xlabel('Fitted values')
axes[0].set_ylabel('Residuals')
axes[0].legend()

# 2) Scale-Location plot (spread vs fitted)
axes[1].scatter(y_hat, np.sqrt(np.abs(std_resid)), alpha=0.2, s=10)
axes[1].set_title('Scale-Location Plot')
axes[1].set_xlabel('Fitted values')
axes[1].set_ylabel('Sqrt(|Standardized Residuals|)')

plt.tight_layout()
plt.savefig("residual_diagnostics.png", dpi=300)  # Save the figure for reference
plt.show()

# Formal diagnostics - Using ONLY Breusch-Pagan (Memory Safe)
bp_lm, bp_lm_pvalue, bp_f, bp_f_pvalue = het_breuschpagan(resid, X_mlr_const)

print("Non-linearity diagnostics")
print(f"Quadratic residual trend coefficient (x^2): {quad_coef[0]:.6f}")

print("\nHeteroscedasticity diagnostics")
print(f"Breusch-Pagan LM p-value: {bp_lm_pvalue:.6g}")
print(f"Breusch-Pagan F p-value : {bp_f_pvalue:.6g}")

# Additional practical signal for funnel-shape
corr_abs_resid_fit = np.corrcoef(np.abs(resid), y_hat)[0, 1]
print(f"Corr(|residual|, fitted): {corr_abs_resid_fit:.4f}")
```
