```python
import numpy as np

# Ensure df is defined from your original data X

# ==============================================================================
# 1. DATETIME & TEMPORAL FEATURES
# ==============================================================================
df['searchDate'] = pd.to_datetime(df['searchDate'])
df['flightDate'] = pd.to_datetime(df['flightDate'])

df['flight_month'] = df['flightDate'].dt.month
df['flight_day_of_week'] = df['flightDate'].dt.dayofweek
df['is_weekend'] = df['flight_day_of_week'].apply(lambda x: 1 if x >= 5 else 0)
df['is_holiday_season'] = df['flight_month'].isin([7, 11, 12]).astype(int)

df['flight_month_sin'] = np.sin(2 * np.pi * df['flight_month'] / 12.0)
df['flight_month_cos'] = np.cos(2 * np.pi * df['flight_month'] / 12.0)

# ==============================================================================
# 2. SEGMENT COLUMN EXTRACTION (RE-FIXED FOR YOUR FORMAT)
# ==============================================================================
# Robust extraction function to handle '2022-05-15T01:35:00.000000000' and '||'
def get_first_hour(val):
    try:
        # Convert to string, take the first segment before '||',
        # then split by 'T' and take the first 2 characters of the time part
        return float(str(val).split('||')[0].split('T')[1][:2])
    except:
        return np.nan

df['journey_start_hour'] = df['segmentsDepartureTime'].apply(get_first_hour)

# Extract Primary Airline
df['primary_airline'] = df['segmentsAirlineCode'].astype(str).str.split('||').str[0]
df['primary_airline'] = df['primary_airline'].replace('nan', np.nan)

# Check for mixed itineraries
df['is_mixed_itinerary'] = df['segmentsAirlineCode'].astype(str).apply(
    lambda x: 1 if len(set(x.split('||'))) > 1 else 0
)

# One-Hot Encode
df = pd.get_dummies(df, columns=['primary_airline'], prefix='airline', drop_first=False)

# ==============================================================================
# 3. TIME-WINDOWED TARGET ENCODING
# ==============================================================================
df['days_to_flight_bin'] = pd.cut(df['numDaysToFlight'],
                                  bins=[0, 7, 14, 30, 60, 365],
                                  labels=['0-7', '8-14', '15-30', '31-60', '61+'],
                                  include_lowest=True)

cols_to_overwrite = [
    'start_airport_month_avg_fare', 'start_airport_month_count',
    'start_airport_daysbin_avg_fare', 'start_airport_daysbin_count',
    'dest_airport_month_avg_fare', 'dest_airport_month_count',
    'dest_airport_daysbin_avg_fare', 'dest_airport_daysbin_count'
]
df = df.drop(columns=[c for c in cols_to_overwrite if c in df.columns], errors='ignore')

if 'startingAirport' in df.columns:
    airport_month_targets = df.groupby(['startingAirport', 'flight_month'], observed=True)['totalFare'].agg(['mean', 'count']).reset_index()
    airport_month_targets.columns = ['startingAirport', 'flight_month', 'start_airport_month_avg_fare', 'start_airport_month_count']
    df = df.merge(airport_month_targets, on=['startingAirport', 'flight_month'], how='left')

    airport_days_targets = df.groupby(['startingAirport', 'days_to_flight_bin'], observed=True)['totalFare'].agg(['mean', 'count']).reset_index()
    airport_days_targets.columns = ['startingAirport', 'days_to_flight_bin', 'start_airport_daysbin_avg_fare', 'start_airport_daysbin_count']
    df = df.merge(airport_days_targets, on=['startingAirport', 'days_to_flight_bin'], how='left')

if 'destinationAirport' in df.columns:
    dest_month_targets = df.groupby(['destinationAirport', 'flight_month'], observed=True)['totalFare'].agg(['mean', 'count']).reset_index()
    dest_month_targets.columns = ['destinationAirport', 'flight_month', 'dest_airport_month_avg_fare', 'dest_airport_month_count']
    df = df.merge(dest_month_targets, on=['destinationAirport', 'flight_month'], how='left')

    dest_days_targets = df.groupby(['destinationAirport', 'days_to_flight_bin'], observed=True)['totalFare'].agg(['mean', 'count']).reset_index()
    dest_days_targets.columns = ['destinationAirport', 'days_to_flight_bin', 'dest_airport_daysbin_avg_fare', 'dest_airport_daysbin_count']
    df = df.merge(dest_days_targets, on=['destinationAirport', 'days_to_flight_bin'], how='left')

# ==============================================================================
# 4. FINAL CLEANUP
# ==============================================================================
y = df['totalFare'].copy()

drop_cols = [
    'legId', 'searchDate', 'flightDate', 'segmentsDepartureTime', 'segmentsArrivalTime',
    'segmentsAirlineName', 'segmentsAirlineCode', 'segmentsAircraft',
    'segmentsArrivalAirportCode', 'segmentsDepartureAirportCode',
    'startingAirport', 'destinationAirport', 'segmentsCabinCode',
    'baseFare', 'totalFare', 'flight_month', 'flight_day_of_week', 'days_to_flight_bin'
]

X_clean = df.drop(columns=[c for c in drop_cols if c in df.columns], errors='ignore')

# Handle types and NaNs
bool_cols = X_clean.select_dtypes(include='bool').columns
X_clean[bool_cols] = X_clean[bool_cols].astype(int)

X_numeric = X_clean.select_dtypes(include=['number'])
X_numeric = X_numeric.replace([np.inf, -np.inf], np.nan)
X_final = X_numeric.fillna(X_numeric.median(numeric_only=True))

print("PIPELINE COMPLETE")
print(f"Features: {X_final.shape[1]} | Rows: {X_final.shape[0]}")
```

##**Synergy part**

```python

X_final['demand_pressure'] = (1 / (X_final['numDaysToFlight'] + 1)) * (1 / (X_final['seatsRemaining'] + 1))

X_final['layover_discomfort_cost'] = X_final['layoverDurationMinutes'] * X_final['travelDuration']

X_final['seasonal_demand_impact'] = X_final['start_airport_month_avg_fare'] * X_final['is_holiday_season']

print("Synergistic features added to X_final.")
```

```python

departure_minutes_from_midnight = X_final['journey_start_hour'] * 60

arrival_minutes_raw = departure_minutes_from_midnight + X_final['travelDuration']

X_final['arrival_minutes_of_day'] = arrival_minutes_raw % (24 * 60)

X_final['arrival_hour'] = X_final['arrival_minutes_of_day'] // 60

late_night_start = 0
late_night_end = 5

X_final['is_late_night_arrival'] = ((X_final['arrival_hour'] >= late_night_start) &
                                    (X_final['arrival_hour'] <= late_night_end)).astype(int)

X_final['late_night_arrival_impact'] = X_final['is_late_night_arrival'] * X_final['travelDuration']

print("New synergy features related to arrival time added to X_final:")
print("  - arrival_hour")
print("  - is_late_night_arrival")
print("  - late_night_arrival_impact")
```

```python
X_final['travel_efficiency'] = X_final['travelDuration'] / (X_final['totalTravelDistance'] + 1)
X_final['layover_proportion'] = X_final['layoverDurationMinutes'] / (X_final['travelDuration'] + 1)
X_final['nonstop_premium_demand'] = X_final['isNonStop'] * X_final['demand_pressure']
```
